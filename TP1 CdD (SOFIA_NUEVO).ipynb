```python
# VARIABLES A USAR
variables = ["ch04","ch06","ch07","ch08","nivel_ed","estado",
             "cat_inac","ipcf","ch09","cat_ocup","ch15",
             "itf","ch03","p47t","p21"]

# FILTRAR REGION PAMPEANA
EPH2005_filtrado = EPH2005[EPH2005['region'] == 'Pampeana'][variables].copy()
EPH2025_filtrado = EPH2025[EPH2025['region'] == 43][variables].copy()

# Convertir nombres de columnas a minúscula
EPH2005_filtrado.columns = EPH2005_filtrado.columns.str.lower()
EPH2025_filtrado.columns = EPH2025_filtrado.columns.str.lower()

# Columnas enteras y float
int_cols = ["ch04","ch03","ch06","ch07","ch08","ch09","ch15",
            "nivel_ed","estado","cat_inac","cat_ocup"]
float_cols = ["ipcf","itf","p47t","p21"]

# FUNCION PARA CONVERTIR A INT64

def convertir_a_int64(df, columnas):
    for col in columnas:
        # Si no es numérica, convertir usando códigos de categoría
        if not pd.api.types.is_numeric_dtype(df[col]):
            df[col] = df[col].astype('category').cat.codes + 1
        df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')
    return df

# Convertir enteros
EPH2005_filtrado = convertir_a_int64(EPH2005_filtrado, int_cols)
EPH2025_filtrado = convertir_a_int64(EPH2025_filtrado, int_cols)

# Convertir floats
for col in float_cols:
    EPH2005_filtrado[col] = pd.to_numeric(EPH2005_filtrado[col], errors='coerce').astype(float)
    EPH2025_filtrado[col] = pd.to_numeric(EPH2025_filtrado[col], errors='coerce').astype(float)

# Revisar tipos y datos faltantes
print("EPH2005 (Región Pampeana):")
print(EPH2005_filtrado.dtypes)
print(EPH2005_filtrado.isna().sum())

print("\nEPH2025 (Región Pampeana):")
print(EPH2025_filtrado.dtypes)
print(EPH2025_filtrado.isna().sum())
```
